```python
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
import numpy as np
import pandas as pd
```

## Loading data

```python
train_data=pd.read_csv('0000000000002429_training_titanic_x_y_train.csv',delimiter=",")
train_data
```

```python
test_data=pd.read_csv("0000000000002429_test_titanic_x_test.csv",delimiter=",")
test_data
```

## Data Preprocessing and Cleaning

```python
title_in_train=[]
for i in train_data['Name']:
    title_in_train.append(i.split()[1][0:-1])
    
title_in_train = np.array(title_in_train)

title_in_train[title_in_train == "Mr"]=0
title_in_train[title_in_train == "Mrs"]=1
title_in_train[title_in_train == "Miss"]=2
title_in_train[title_in_train == "Master"]=3
title_in_train[(title_in_train != '0')&(title_in_train != '1')&(title_in_train != '2')&(title_in_train != '3')]=4
title_in_train = np.array(title_in_train,dtype =int)
title_in_train
```

```python
title_in_test = []

for i in test_data['Name']:
    title_in_test.append(i.split()[1][0:-1])
    
title_in_test = np.array(title_in_test)

title_in_test[title_in_test == "Mr"]=0
title_in_test[title_in_test == "Mrs"]=1
title_in_test[title_in_test == "Miss"]=2
title_in_test[title_in_test == "Master"]=3
title_in_test[(title_in_test != '0')&(title_in_test != '1')&(title_in_test != '2')&(title_in_test != '3')]=4
title_in_test = np.array(title_in_test,dtype =int)
title_in_test
```

```python
train_data.drop('Name',axis=1,inplace=True)
train_data.drop('Ticket',axis=1,inplace=True)
```

```python
train_data
```

```python
test_data.drop('Name',axis=1,inplace=True)
test_data.drop('Ticket',axis=1,inplace=True)
```

```python
test_data
```

```python
def gen(s):
    if s == "male":
        return 0
    elif s == "female":
        return 1
```

```python
train_data["Sex"] = train_data["Sex"].apply(gen)
train_data.head(3)
```

```python
train_data.isnull().sum()
```

```python
train_data["Age"].fillna(train_data["Age"].mean(),inplace = True)
train_data.isnull().sum()
```

```python
train_data.loc[train_data['Cabin'].isnull(),'Cabin']=0
train_data.loc[train_data['Cabin']!=0,'Cabin']=1
```

```python
train_data.isnull().sum()
```

```python
# let s=1,Q=2,C=3
def emba(x):
    if x=='S':
        return 1
    elif x=='Q':
        return 2
    else:
        return 3
```

```python
train_data["Embarked"]=train_data['Embarked'].apply(emba)
```

```python
train_data
```

```python
train_data.describe()
```

```python
training_data=train_data.values
x_train=training_data[:,:-1]
y_train=training_data[:,-1]
x_train=np.append(x_train,title_in_train.reshape(-1,1),axis=1)
```

```python
x_train.shape
```

```python
y_train=y_train.astype('int')
```

```python
x_train=x_train.astype('int')
```

```python
x_train
```

```python
# working on testing data
# for sex
test_data["Sex"] = test_data["Sex"].apply(gen)
# for embarked
test_data["Embarked"]=test_data['Embarked'].apply(emba)
# for age
test_data["Age"].fillna(test_data["Age"].mean(),inplace = True)
# for cabin
test_data.loc[test_data['Cabin'].isnull(),'Cabin']=0
test_data.loc[test_data['Cabin']!=0,'Cabin']=1
test_data
```

```python
test_data=test_data.values
test_data=np.append(test_data,title_in_test.reshape(-1,1),axis=1)
test_data.shape
```

# Applying Logistic Regression algorithm for the given data

```python
clf = LogisticRegression(solver='saga',max_iter=10000000, tol=0.00001)
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(test_data)
```

```python
np.savetxt('logistic_regression_asgnmt.csv',y_pred,delimiter=",")
```
